```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nltk
```

```python
url = "https://en.wikipedia.org/wiki/Lagos#Demography"
```

```python
tables = pd.read_html(url)
```

```python
len(tables)
```

```python
demography = tables[2]
```

```python
len(demography)
```

```python
demography.drop(index=8, inplace=True)
```

```python
print(demography.head())
```

```python
demography_year = demography['Year']
demography_pop = demography['Pop.']
```

```python
demography_pop
```

```python
plt.title('Lagos State Population Growth from 1950-2019')
plt.plot(demography_year, demography_pop, 'ForestGreen')
plt.ylabel('Population in millions')
plt.xlabel('Years')
plt.margins(0.002)
#plt.subplots_adjust(bottom=0.15)
plt.grid(color="lightslategrey", alpha=0.15, zorder=1)

plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

plt.show()
```
