```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('boston_housing.csv')
crim = data.iloc[:, 0].values
zn = data.iloc[:, 1].values
indus = data.iloc[:, 2].values
nox = data.iloc[:, 4].values
rm = data.iloc[:, 5].values
age = data.iloc[:, 6].values
y = data.iloc[:, -1].values
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(2, 3, 1)
plt.scatter(crim, y, color='blue', label='CRIM')
plt.legend()

plt.subplot(2, 3, 2)
plt.scatter(zn, y, color='red', label='ZN')
plt.legend()

plt.subplot(2, 3, 3)
plt.scatter(indus, y, color='yellow', label='INDUS')
plt.legend()

plt.subplot(2, 3, 4)
plt.scatter(nox, y, color='green', label='NOX')
plt.legend()

plt.subplot(2, 3, 5)
plt.scatter(rm, y, color='orange', label='RM')
plt.legend()

plt.subplot(2, 3, 6)
plt.scatter(age, y, color='k', label='AGE')
plt.legend()

plt.suptitle("Boston Housing Prices")
plt.show()
```

```python
score = [100, 25, 75, 80, 92]
poets = ["Arthur Rimbaud", "Niyi Osundare", "Christopher Okigbo", "Sylva Plath", "Okogbule Wonodi"]
the_explode = [0.1, 0, 0, 0, 0]
poet_colors = ['yellow', 'lightblue', 'orange', 'lightgreen', 'brown']

plt.pie(score, labels=poets, explode=the_explode, shadow=True, colors=poet_colors)
plt.legend(title='My Favorite Poets', loc='lower right', bbox_to_anchor=(2, 0, 0, 0))
plt.show()
```
