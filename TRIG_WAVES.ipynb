```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from matplotlib import markers
x_axis = np.arange(0, 10, 0.1)
sin_y = np.sin(x_axis)
cos_y = np.cos(x_axis)
tan_y = np.tan(x_axis)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
line1, =plt.plot(x_axis, sin_y, color='red', label='Sine Wave')
plt.title('Sine Wave')

plt.subplot(1, 3, 2)
line2, =plt.plot(x_axis, cos_y, color='blue', label='Cosine Wave')
plt.title('Cosine Wave')

plt.subplot(1, 3, 3)
line3, =plt.plot(x_axis, tan_y, color='orange', label='Tangent Wave')
plt.title('Tangent Wave')

plt.legend([line1, line2, line3], ['Sine Wave', 'Cosine Wave', 'Tangent Wave'], loc="lower right", bbox_to_anchor=(0.5, 0, 1, 2))

plt.show()

```
