```python
X = data.iloc[:, 1:-1].values
y = data.iloc[:, -1].values
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()

# Encode the "sex" column
X[:, :1] = np.array(encoder.fit_transform(X[:, :1]))

# Encode the "address" column
X[:, 2:3] = np.array(encoder.fit_transform(X[:, 2:3]))

# Encode the "famsize" column
X[:, 3:4] = np.array(encoder.fit_transform(X[:, 3:4]))

# Encode the "pstatus" column
X[:, 4:5] = np.array(encoder.fit_transform(X[:, 4:5]))

# Encode the "Mjob" and "Fjob" column
X[:, 7:11] = np.array(encoder.fit_transform(X[:, 7:11]))

# Encode the  column
X[:, 14:22] = np.array(encoder.fit_transform(X[:, 14:22]))
```

```python
print(X[0:10])
```

```python
print(X[0:20, :])
```

```python
df = {
    "sex": X[:, 0:1].reshape(-1),
    "age": X[:, 1:2].reshape(-1),
    "address": X[:, 2:3].reshape(-1),
    "score": X[:, -1].reshape(-1)
}
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1)
```

```python
from sklearn.metrics import r2_score
r2_score(y_pred, y_test)
```
